```python
import pandas as pd
```

```python
# Read the Excel file as a DataFrame
df_full = pd.read_excel("survey.xlsx")
print("Num rows:", len(df_full))
df_full.head()
```

```python
# Filter the DataFrame to only include rows where the "Finished" column is "True"
df_scores = df_full[df_full["Finished"] == "True"].copy()
print("Num rows:", len(df_scores))

rating_map = {
    "Strongly disagree": 1,
    "Disagree": 2,
    "Somewhat disagree": 3,
    "Somewhat agree": 4,
    "Agree": 5,
    "Strongly agree": 6
}

# Convert text responses to numeric where applicable
for col in df_scores.columns:
    if col.startswith("Q10_"):
        df_scores.loc[:, col] = df_scores[col].map(rating_map)

df_scores.head()
```

```python
emotional_regulation_stages = {
    "Confront_Unpleasant_Situations": ["Q10_9", "Q10_11", "Q10_15", "Q10_28"],
    "Avoid_Unpleasant_Situations":    ["Q10_4", "Q10_5", "Q10_12", "Q10_18", "Q10_42", "Q10_44"],
    "Resolve_Conflicts":              ["Q10_6", "Q10_17", "Q10_19", "Q10_26", "Q10_36", "Q10_39"],
    "Avoid_Conflicts":                ["Q10_3", "Q10_10", "Q10_14", "Q10_37", "Q10_43"],
    "Focus_Elsewhere":                ["Q10_2", "Q10_22", "Q10_23", "Q10_31"],
    "Cognitively_Distract":           ["Q10_13", "Q10_27", "Q10_29", "Q10_33", "Q10_35"],
    "Consider_Benefits":              ["Q10_7", "Q10_20", "Q10_21", "Q10_24", "Q10_40", "Q10_45"],
    "Reduce_Importance":              ["Q10_25", "Q10_38", "Q10_41"],
    "Support_by_Emotion_Sharing":     ["Q10_8", "Q10_30", "Q10_32"],
    "Expressive_Suppression":         ["Q10_1", "Q10_16", "Q10_34"]
}

# Calculate the mean for each emotional regulation stage
for emotional_stage, questions in emotional_regulation_stages.items():
    df_scores[emotional_stage] = df_scores[questions].mean(axis=1, skipna=True)
    
df_scores.head()
```

Build demographics table (44 rows, dashes for missing)

```python
# Download the DataFrame as a CSV file
df_scores.to_csv("pmerq-survey-scores.csv", index=False)
```

```python
# After reading df_full, drop that unwanted “extra” second row:
df_full = pd.read_excel("survey.xlsx")
df_full = df_full.reset_index(drop=True)
df_full = df_full.drop(df_full.index[0]).reset_index(drop=True)

demo_cols = {
    "ResponseId": "ParticipantID",
    "Q5":           "Age Range",
    "Q6":           "Gender",
    "Q7":           "Race",
    "Q8":           "Education",
    "Q9":           "Familiarity",
    "Group":        "Group",
    "Device":       "Device"
}

df_demo = df_full[list(demo_cols)].rename(columns=demo_cols)

# Clean the Familiarity text and only retain the first character
df_demo["Familiarity"] = df_demo["Familiarity"].astype(str).str[0]

# Collapse any “Other (please specify)” to just “Other”
df_demo["Race"] = (
    df_demo["Race"]
      .astype(str)
      .str.replace(r"(?i)^Other.*$", "Other", regex=True)
)

# Fill any NaN with dash
df_demo = df_demo.fillna("-")

# print("Participants in demo table:", len(df_demo) - 1)  # should be 43

latex = df_demo.to_latex(
    index=False,
    na_rep="-",            # simple hyphen
    caption="Participant demographics",
    label="tab:demographics",
    escape=True            # <— turn escaping back on
)

print(latex)
```

Overall summary stats (mean ± SD) for each emotional stage

```python
# only on finished participants
raw = df_scores[list(emotional_regulation_stages)].agg(["mean", "std"]).T

# create the single formatted column
raw["Mean ± SD"] = raw.apply(
    lambda r: f"{r['mean']:.2f} \u00B1 {r['std']:.2f}", axis=1
)

# keep only that combined column
stats_single = raw[["Mean ± SD"]]

print(stats_single.to_latex(
    index=True,
    header=["Mean \\pm SD"],    # if you want to LaTeX-escape the ±
    escape=True,
    caption="Overall mean ± SD by emotional regulation stage"
))
```

Group‑level summary (Groups 1–12)

```python
# 1. Filter groups 1–12
df_g1_12 = df_scores[df_scores["Group"].astype(float).between(1, 12)]

# 2. Compute the raw means and stds
raw = (
    df_g1_12
    .groupby("Group")[list(emotional_regulation_stages)]
    .agg(["mean", "std"])
)

# 3. Build a new DataFrame where each stage is one column formatted as "M ± S"
grp_stats_single = pd.DataFrame(index=raw.index)
for stage in emotional_regulation_stages:
    mean_col = raw[(stage, "mean")]
    std_col  = raw[(stage, "std")]
    grp_stats_single[stage] = (
        mean_col.map("{:.2f}".format)
        + " ± "
        + std_col.map("{:.2f}".format)
    )

# 4. (Optionally rename columns for prettier labels)
grp_stats_single.columns = [stage.replace("_", " ") for stage in grp_stats_single.columns]

# 5. Export to LaTeX
print(grp_stats_single.to_latex(
    index=True,
    caption="Mean \\(\\pm\\) SD by group (1–12)",
    escape=False
))
```
