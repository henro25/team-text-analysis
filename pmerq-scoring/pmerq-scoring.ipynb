```python
import pandas as pd
```

```python
# Read the Excel file as a DataFrame
df = pd.read_excel("survey.xlsx")
print("Num rows:", len(df))
df.head()
```

```python
# Filter the DataFrame to only include rows where the "Finished" column is "True"
df_finished = df[df["Finished"] == "True"].copy()
print("Num rows:", len(df_finished))

rating_map = {
    "Strongly disagree": 1,
    "Disagree": 2,
    "Somewhat disagree": 3,
    "Somewhat agree": 4,
    "Agree": 5,
    "Strongly agree": 6
}

# Convert text responses to numeric where applicable
for col in df_finished.columns:
    if col.startswith("Q10_"):
        df_finished.loc[:, col] = df_finished[col].map(rating_map)

df_finished.head()
```

```python
emotional_regulation_stages = {
    "Confront_Unpleasant_Situations": ["Q10_9", "Q10_11", "Q10_15", "Q10_28"],
    "Avoid_Unpleasant_Situations":    ["Q10_4", "Q10_5", "Q10_12", "Q10_18", "Q10_42", "Q10_44"],
    "Resolve_Conflicts":              ["Q10_6", "Q10_17", "Q10_19", "Q10_26", "Q10_36", "Q10_39"],
    "Avoid_Conflicts":                ["Q10_3", "Q10_10", "Q10_14", "Q10_37", "Q10_43"],
    "Focus_Elsewhere":                ["Q10_2", "Q10_22", "Q10_23", "Q10_31"],
    "Cognitively_Distract":           ["Q10_13", "Q10_27", "Q10_29", "Q10_33", "Q10_35"],
    "Consider_Benefits":              ["Q10_7", "Q10_20", "Q10_21", "Q10_24", "Q10_40", "Q10_45"],
    "Reduce_Importance":              ["Q10_25", "Q10_38", "Q10_41"],
    "Support_by_Emotion_Sharing":     ["Q10_8", "Q10_30", "Q10_32"],
    "Expressive_Suppression":         ["Q10_1", "Q10_16", "Q10_34"]
}

# Calculate the mean for each emotional regulation stage
for emotional_stage, questions in emotional_regulation_stages.items():
    df_finished[emotional_stage] = df_finished[questions].mean(axis=1, skipna=True)
    
df_finished.head()
```

```python
# Download the DataFrame as a CSV file
df_finished.to_csv("pmerq-survey-scores.csv", index=False)
```
